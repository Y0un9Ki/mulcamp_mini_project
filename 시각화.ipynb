```python
import pandas as pd
import plotly.express as px
```

```python
df_김포단독다가구=pd.read_csv('data/df_김포단독다가구.csv')
df_김포아파트=pd.read_csv('data/df_김포아파트.csv')
df_김포오피스텔=pd.read_csv('data/df_김포오피스텔.csv')
df_김포연립다세대=pd.read_csv('data/df_김포연립다세대.csv')
```

```python
df_김포단독다가구['계약년월']= df_김포단독다가구["계약년월"].astype(str)
df_김포아파트['계약년월']= df_김포아파트["계약년월"].astype(str)
df_김포연립다세대['계약년월']= df_김포연립다세대["계약년월"].astype(str)
df_김포오피스텔['계약년월']= df_김포오피스텔["계약년월"].astype(str)
```

```python
df_김포단독다가구=df_김포단독다가구.sort_values(by='계약년월', ascending=True, ignore_index=True)
df_김포아파트=df_김포아파트.sort_values(by='계약년월', ascending=True, ignore_index=True)
df_김포오피스텔=df_김포오피스텔.sort_values(by='계약년월', ascending=True, ignore_index=True)
df_김포연립다세대=df_김포연립다세대.sort_values(by='계약년월', ascending=True, ignore_index=True)
```

```python
df_김포단독다가구=df_김포단독다가구.assign(주택유형='단독다가구')
df_김포아파트=df_김포아파트.assign(주택유형='아파트')
df_김포오피스텔=df_김포오피스텔.assign(주택유형='오피스텔')
df_김포연립다세대=df_김포연립다세대.assign(주택유형='연립다세대')
```

```python
df_김포=pd.concat([df_김포단독다가구,df_김포아파트,df_김포오피스텔,df_김포연립다세대])
```

```python
df_김포['시군구'].value_counts()
```

```python
김포_copy= df_김포['시군구'].unique()
```

```python
for i in 김포_copy:
    df_김포=df_김포.replace({i : i[8:11]})
```

```python
df_김포=df_김포.sort_values(by='계약년월',ascending=True, ignore_index=True)
```

```python
df_김포_발표후=df_김포.loc[df_김포['계약년월']>='202310', :].assign(발표='발표후')
```

```python
df_김포_발표전=df_김포.loc[(df_김포['계약년월']<'202310') & ((df_김포['계약년월']>='202306')), :].assign(발표='발표전')
```

```python
df_김포_발표전
```

```python
df_김포=pd.concat([df_김포_발표전,df_김포_발표후])
```

```python
df_김포.to_csv("data/df_김포.csv", index=False)
```

```python
df_서울단독다가구=pd.read_csv('data/df_서울단독다가구.csv')
df_서울아파트=pd.read_csv('data/df_서울아파트.csv')
df_서울연립다세대=pd.read_csv('data/df_서울연립다세대.csv')
df_서울오피스텔=pd.read_csv('data/df_서울오피스텔.csv')
```

```python
df_서울오피스텔['계약년월']= df_서울오피스텔["계약년월"].astype(str)
df_서울아파트['계약년월']= df_서울아파트["계약년월"].astype(str)
df_서울단독다가구['계약년월']= df_서울단독다가구["계약년월"].astype(str)
df_서울연립다세대['계약년월']= df_서울연립다세대["계약년월"].astype(str)
```

```python
df_서울단독다가구=df_서울단독다가구.loc[df_서울단독다가구["계약년월"]>='202306',:]
df_서울아파트=df_서울아파트.loc[df_서울아파트["계약년월"]>='202306',:]
df_서울연립다세대=df_서울연립다세대.loc[df_서울연립다세대["계약년월"]>='202306',:]
df_서울오피스텔=df_서울오피스텔.loc[df_서울오피스텔["계약년월"]>='202306',:]
```

```python
df_서울단독다가구=df_서울단독다가구.assign(주택유형='단독다가구')
df_서울아파트=df_서울아파트.assign(주택유형='아파트')
df_서울연립다세대=df_서울연립다세대.assign(주택유형='연립다세대')
df_서울오피스텔=df_서울오피스텔.assign(주택유형='오피스텔')
```

```python
df_서울=pd.concat([df_서울단독다가구,df_서울아파트,df_서울연립다세대,df_서울오피스텔])
```

```python
df_서울.sort_values(by='계약년월',ascending=True, ignore_index=True)
```

```python
df_서울B=df_서울.loc[(df_서울['계약년월']>='202306') & (df_서울['계약년월']<'202310'), :].assign(발표='발표전').sort_values(by='계약년월',ascending=True, ignore_index=True)
```

```python
df_서울A=df_서울.loc[df_서울['계약년월']>='202310', :].assign(발표='발표후').sort_values(by='계약년월',ascending=True, ignore_index=True)
```

```python
df_서울=pd.concat([df_서울B,df_서울A])
```
